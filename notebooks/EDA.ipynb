```python
# EDA.ipynb - Exploratory Data Analysis for Chest X-ray Disease Dataset

# Purpose:
# This script performs Exploratory Data Analysis (EDA) to understand the structure,
# distribution, and balance of the dataset before training a deep learning model.

# Main Tasks Performed:
# 1. Displays sample images from each class
# 2. Counts the number of samples per class
# 3. Visualizes class distribution
# 4. Checks image shape, size consistency
# 5. Optional: Displays basic statistics or image intensity histogram
```

```python

import os
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.preprocessing.image import load_img
```

```python

# Define dataset paths
DATASET_PATH = '../data/raw'
CLASSES = ['Covid', 'Normal', 'Pneumonia', 'Tuberculosis']
```

```python

# Count images per class
class_counts = {}
for cls in CLASSES:
    class_dir = os.path.join(DATASET_PATH, cls)
    class_counts[cls] = len(os.listdir(class_dir))
```

```python

# Plot class distribution
plt.figure(figsize=(8, 6))
sns.barplot(x=list(class_counts.keys()), y=list(class_counts.values()), palette='viridis')
plt.title('Class Distribution')
plt.ylabel('Number of Images')
plt.xticks(rotation=45)
plt.tight_layout()
#plt.savefig('../reports/class_distribution.png')
plt.show()
```

```python

# Show sample images from each class
plt.figure(figsize=(12, 8))
for i, cls in enumerate(CLASSES):
    img_path = os.path.join(DATASET_PATH, cls, os.listdir(os.path.join(DATASET_PATH, cls))[0])
    img = load_img(img_path, target_size=(224, 224))
    plt.subplot(2, 2, i+1)
    plt.imshow(img)
    plt.title(cls)
    plt.axis('off')

plt.tight_layout()
#plt.savefig('../reports/sample_images.png')
plt.show()

print("EDA complete. Visualizations saved in 'reports/' folder.")
```
